```python
import arcpy
import os
import shutil
import time
from datetime import timedelta
from arcpy.sa import Raster, SetNull, CellStatistics, Con

# -----------------------------
# Setup environment
# -----------------------------
arcpy.env.workspace = r"C:\Users\ss2596\Documents\SENTINEL2 Mufunta\Mufunta_Sentinel2_6_2025"
arcpy.env.overwriteOutput = True

# -----------------------------
# Sentinel-2 RGB bands
# -----------------------------
bands = {"B02": "Blue", "B03": "Green", "B04": "Red"}  # band codes
parent_folder = arcpy.env.workspace
temp_folder = os.path.join(parent_folder, "temp_tiffs")
os.makedirs(temp_folder, exist_ok=True)

mosaicked_band_files = []
threshold = 1000  # pixels below this value are considered invalid

# -----------------------------
# Collect SAFE folders alphabetically (priority)
# -----------------------------
safe_folders = sorted([f for f in os.listdir(parent_folder) if f.endswith(".SAFE")])
print(f"Found {len(safe_folders)} SAFE folders to process")
print(f"Will process {len(bands)} bands: {', '.join(bands.values())}")
print(f"Expected total rasters to process: ~{len(safe_folders) * len(bands)}")

# Initialize timing variables
start_time = time.time()
total_rasters_expected = len(safe_folders) * len(bands)
rasters_processed = 0

def format_time(seconds):
    """Convert seconds to human readable format"""
    if seconds < 60:
        return f"{seconds:.0f}s"
    elif seconds < 3600:
        return f"{seconds//60:.0f}m {seconds%60:.0f}s"
    else:
        return f"{seconds//3600:.0f}h {(seconds%3600)//60:.0f}m"

def get_eta_info(processed, total, elapsed_time):
    """Calculate ETA and processing speed"""
    if processed == 0:
        return "calculating...", "calculating..."
    
    rate = processed / elapsed_time  # rasters per second
    remaining = total - processed
    eta_seconds = remaining / rate if rate > 0 else 0
    
    speed_info = f"{rate*60:.1f}/min" if rate > 0 else "calculating..."
    eta_info = format_time(eta_seconds) if eta_seconds > 0 else "calculating..."
    
    return speed_info, eta_info

print(f"\nâ±ï¸  Processing started at {time.strftime('%H:%M:%S')}")

# -----------------------------
# Process each band
# -----------------------------
for band_idx, (band_code, band_name) in enumerate(bands.items(), 1):
    print(f"\n[{band_idx}/{len(bands)}] Processing {band_name} ({band_code})...")

    tiff_list = []
    counter = 0

    # Process all SAFE folders
    for safe_idx, safe in enumerate(safe_folders, 1):
        granule_path = os.path.join(parent_folder, safe, "GRANULE")
        if not os.path.exists(granule_path):
            continue
        
        for granule_id in os.listdir(granule_path):
            img_data_path = os.path.join(granule_path, granule_id, "IMG_DATA", "R10m")
            if not os.path.exists(img_data_path):
                continue

            # Find band JP2 files
            for f in os.listdir(img_data_path):
                if band_code in f and f.endswith(".jp2") and not f.startswith("MSK_"):
                    counter += 1
                    jp2_path = os.path.join(img_data_path, f)
                    
                    # Create unique filenames
                    unique_id = f"{counter:03d}_{granule_id[:8]}"
                    tiff_path = os.path.join(temp_folder, f"{unique_id}_{band_code}.tif")
                    masked_tiff = os.path.join(temp_folder, f"{unique_id}_{band_code}_masked.tif")
                    
                    # Clean up existing files
                    for file_path in [tiff_path, masked_tiff]:
                        if os.path.exists(file_path):
                            try:
                                os.remove(file_path)
                            except:
                                pass
                    
                    # Update timing and show progress with ETA
                    rasters_processed += 1
                    elapsed_time = time.time() - start_time
                    progress_pct = (safe_idx / len(safe_folders)) * 100
                    speed, eta = get_eta_info(rasters_processed, total_rasters_expected, elapsed_time)
                    
                    print(f"  Processing raster {counter} ({progress_pct:.0f}% complete) - "
                          f"Elapsed: {format_time(elapsed_time)} | Speed: {speed} | ETA: {eta}")
                    
                    # Process the file
                    arcpy.management.CopyRaster(jp2_path, tiff_path)
                    band_r = Raster(tiff_path)
                    
                    # More comprehensive masking: exclude invalid pixels (0, very low values, and NoData)
                    # Also check for typical Sentinel-2 fill values
                    masked_r = SetNull((band_r <= threshold) | (band_r == 0) | (band_r >= 65535), band_r)
                    arcpy.management.CopyRaster(masked_r, masked_tiff)
                    tiff_list.append(masked_tiff)
                    
                    # Clean up
                    del band_r, masked_r
                    try:
                        os.remove(tiff_path)
                    except:
                        pass

    if not tiff_list:
        print(f"  No valid rasters found for {band_name}")
        continue

    # -----------------------------
    # Mosaic using MEAN (average pixel values, ignoring masked)
    # -----------------------------
    out_raster = os.path.join(temp_folder, f"mosaic_{band_code}.tif")
    
    # Delete existing mosaic if it exists
    if os.path.exists(out_raster):
        try:
            os.remove(out_raster)
        except:
            pass
    
    print(f"  Found {len(tiff_list)} rasters for {band_name}")
    print(f"  Creating averaged mosaic with strict threshold filtering...")
    
    if len(tiff_list) > 1:
        try:
            print(f"    Setting up environment for {len(tiff_list)} rasters...")
            arcpy.env.extent = "MAXOF"
            arcpy.env.cellSize = "MINOF"
            
            # Clear any existing workspace cache to prevent memory issues
            arcpy.ClearWorkspaceCache_management()
            
            print(f"    Computing mean statistics...")
            # Use MEAN with DATA option to ignore NoData pixels in averaging
            # This ensures proper pixel-wise averaging across all valid pixels
            mean_raster = CellStatistics(tiff_list, "MEAN", "DATA")
            
            print(f"    Applying threshold mask...")
            # Apply strict threshold filtering to ensure no invalid pixels are included
            # This removes any pixels that fall outside our valid range after averaging
            final_masked = SetNull((mean_raster <= threshold) | 
                                 (mean_raster == 0) | 
                                 (mean_raster >= 65535), mean_raster)
            
            print(f"    Saving final mosaic...")
            arcpy.management.CopyRaster(final_masked, out_raster)
            
            # Clean up memory immediately
            del mean_raster, final_masked
            
            # Reset environment
            arcpy.env.extent = None
            arcpy.env.cellSize = None
            
            print(f"    âœ“ Mosaic creation completed successfully")
            
            # Clean up individual files
            for tiff_file in tiff_list:
                try:
                    os.remove(tiff_file)
                except:
                    pass
            
        except Exception as stats_error:
            print(f"  Using fallback mosaic method...")
            arcpy.management.CopyRaster(tiff_list[0], out_raster)
            for i, tiff in enumerate(tiff_list[1:], 1):
                arcpy.management.Mosaic(
                    inputs=[tiff],
                    target=out_raster,
                    mosaic_type="MEAN"
                )
    else:
        arcpy.management.CopyRaster(tiff_list[0], out_raster)

    mosaicked_band_files.append(out_raster)
    print(f"  âœ“ {band_name} mosaic completed")

# -----------------------------
# Combine bands into final RGB raster
# -----------------------------
if mosaicked_band_files:
    rgb_order = [
        os.path.join(temp_folder, "mosaic_B04.tif"),  # Red
        os.path.join(temp_folder, "mosaic_B03.tif"),  # Green
        os.path.join(temp_folder, "mosaic_B02.tif")   # Blue
    ]

    # Use parent folder name for final output
    folder_name = os.path.basename(parent_folder.rstrip("\\/"))
    final_rgb = os.path.join(parent_folder, f"{folder_name}.tif")

    print(f"\nCreating final RGB composite...")
    arcpy.management.CompositeBands(rgb_order, final_rgb)
    
    print("Calculating statistics...")
    arcpy.management.CalculateStatistics(final_rgb)
    print(f"âœ… RGB mosaic saved: {final_rgb}")

    # Final timing summary
    total_elapsed = time.time() - start_time
    print(f"\nâ±ï¸  Processing completed at {time.strftime('%H:%M:%S')}")
    print(f"â±ï¸  Total processing time: {format_time(total_elapsed)}")
    if rasters_processed > 0:
        avg_rate = rasters_processed / total_elapsed
        print(f"â±ï¸  Average processing rate: {avg_rate*60:.1f} rasters/minute")


# -----------------------------
# Delete Temporary Files
# -----------------------------
print(f"\nCleaning up temporary files...")

# Clear variables
try:
    del band_r
except:
    pass
try:
    del masked_r
except:
    pass
try:
    del mean_raster
except:
    pass

# Clear ArcPy cache
arcpy.ClearWorkspaceCache_management()
arcpy.ClearEnvironment("workspace")

import gc, time
gc.collect()
time.sleep(3)

try:
    shutil.rmtree(temp_folder)
    print("âœ… Cleanup completed")
except Exception as e:
    print(f"âš ï¸ Could not remove temp folder: {e}")
```

```python
import arcpy
import os
import numpy as np
from arcpy.sa import Raster

# -----------------------------
# Cloud Coverage Analysis and Folder Renaming Tool
# -----------------------------

# Configuration - Edit these paths and thresholds as needed
base_directory = r"C:\Users\ss2596\Documents\SENTINEL2 Mufunta\Mufunta_Sentinel2_6_2025"
cloud_threshold = 10  # Cloud probability threshold (0-100) for pixel classification
max_cloud_coverage = 1  # Maximum allowed cloud coverage (for reporting only, not filtering)

arcpy.env.overwriteOutput = True
    
# Get all directories that might contain SAFE folders
if not os.path.exists(base_directory):
    print(f"âŒ Directory not found: {base_directory}")
else:
    # Look for directories containing SAFE folders
    folders_to_process = []
    
    # Check if current directory has SAFE folders
    safe_folders = [f for f in os.listdir(base_directory) if f.endswith(".SAFE")]
    if safe_folders:
        folders_to_process.append(base_directory)
    
    # Check subdirectories for SAFE folders
    for item in os.listdir(base_directory):
        item_path = os.path.join(base_directory, item)
        if os.path.isdir(item_path):
            sub_safe_folders = [f for f in os.listdir(item_path) if f.endswith(".SAFE")]
            if sub_safe_folders:
                folders_to_process.append(item_path)
    
    if not folders_to_process:
        print(f"âŒ No folders containing SAFE files found in: {base_directory}")
    else:
        print(f"Found {len(folders_to_process)} folders to analyze:")
        for folder in folders_to_process:
            print(f"  {folder}")
        
        # Process each folder
        for folder_path in folders_to_process:
            folder_name = os.path.basename(folder_path.rstrip("\\/"))
            
            # Skip if already processed (has cloud percentage in name)
            if "_cloud" in folder_name and "pct" in folder_name:
                print(f"â­ï¸  Skipping {folder_name} (already processed)")
                continue
            
            print(f"\nğŸ“Š Analyzing cloud coverage for: {folder_name}")
            
            # Get SAFE folders in this directory
            safe_folders = sorted([f for f in os.listdir(folder_path) if f.endswith(".SAFE")])
            print(f"  Found {len(safe_folders)} SAFE folders")
            
            total_cloud_pixels = 0
            total_pixels = 0
            tiles_analyzed = 0
            
            # Analyze each SAFE folder individually
            for safe in safe_folders:
                safe_folder_path = os.path.join(folder_path, safe)
                
                # Skip if already processed (has cloud percentage prefix or suffix)
                if (safe.startswith("cloud") and "pct_" in safe) or ("_cloud" in safe and "pct" in safe):
                    print(f"  â­ï¸  Skipping {safe} (already processed)")
                    continue
                
                print(f"  ğŸ“Š Analyzing SAFE folder: {safe}")
                
                granule_path = os.path.join(safe_folder_path, "GRANULE")
                if not os.path.exists(granule_path):
                    print(f"    âŒ No GRANULE folder found")
                    continue
                
                safe_cloud_pixels = 0
                safe_total_pixels = 0
                safe_tiles_analyzed = 0
                
                for granule_id in os.listdir(granule_path):
                    qi_data_path = os.path.join(granule_path, granule_id, "QI_DATA")
                    
                    if not os.path.exists(qi_data_path):
                        continue
                    
                    # Look for cloud mask
                    cloud_mask_path = None
                    cloud_mask_type = None
                    
                    qi_files = os.listdir(qi_data_path)
                    for qi_file in qi_files:
                        if "MSK_CLDPRB_60m" in qi_file and qi_file.endswith(".jp2"):
                            cloud_mask_type = "ESA_PROB"
                            cloud_mask_path = os.path.join(qi_data_path, qi_file)
                            break
                        elif "MSK_CLDPRB" in qi_file and qi_file.endswith(".jp2"):
                            cloud_mask_type = "ESA_PROB"
                            cloud_mask_path = os.path.join(qi_data_path, qi_file)
                        elif "CLP" in qi_file and qi_file.endswith(".jp2"):
                            cloud_mask_type = "SENTINELHUB_PROB"
                            cloud_mask_path = os.path.join(qi_data_path, qi_file)
                        elif "CLM" in qi_file and qi_file.endswith(".jp2"):
                            cloud_mask_type = "SENTINELHUB_MASK"
                            cloud_mask_path = os.path.join(qi_data_path, qi_file)
                    
                    if cloud_mask_path and os.path.exists(cloud_mask_path):
                        try:
                            print(f"    Processing {granule_id[:15]}... ", end="")
                            
                            cloud_raster = Raster(cloud_mask_path)
                            tile_pixels = float(cloud_raster.width * cloud_raster.height)
                            
                            # Get max value for threshold adjustment
                            try:
                                max_result = arcpy.GetRasterProperties_management(cloud_raster, "MAXIMUM")
                                max_val = float(max_result.getOutput(0))
                            except:
                                max_val = 100 if cloud_mask_type == "ESA_PROB" else 255
                            
                            # Adjust threshold based on mask type
                            if cloud_mask_type == "SENTINELHUB_PROB":
                                adjusted_threshold = (cloud_threshold / 100.0) * 255
                            elif cloud_mask_type == "SENTINELHUB_MASK":
                                adjusted_threshold = 0.5
                            elif max_val <= 10:
                                adjusted_threshold = (cloud_threshold / 100.0) * max_val
                            else:
                                adjusted_threshold = cloud_threshold
                            
                            # Use official ESA cloud percentage from L2A quality XML file
                            import xml.etree.ElementTree as ET
                            
                            # Look for L2A_QUALITY.xml file in QI_DATA folder
                            quality_xml_path = os.path.join(qi_data_path, "L2A_QUALITY.xml")
                            if not os.path.exists(quality_xml_path):
                                # Try alternative name
                                quality_xml_path = os.path.join(qi_data_path, "l2a_quality.xml")
                            
                            if os.path.exists(quality_xml_path):
                                try:
                                    # Parse XML and extract CLOUDY_PIXEL_PERCENTAGE
                                    tree = ET.parse(quality_xml_path)
                                    root = tree.getroot()
                                    
                                    # Find the CLOUDY_PIXEL_PERCENTAGE value
                                    tile_cloud_pct = None
                                    for elem in root.iter():
                                        if elem.get('name') == 'CLOUDY_PIXEL_PERCENTAGE':
                                            tile_cloud_pct = float(elem.text)
                                            break
                                    
                                    if tile_cloud_pct is not None:
                                        print(f"Official ESA cloud coverage: {tile_cloud_pct:.1f}%")
                                        
                                        # Add to totals (use percentage as weight, assuming equal tile sizes)
                                        safe_cloud_pixels += tile_cloud_pct  # Store percentage directly
                                        safe_total_pixels += 100.0  # Each tile contributes 100% to the average
                                        safe_tiles_analyzed += 1
                                        
                                        print(f"    â””â”€â”€ Using official ESA calculation from L2A_QUALITY.xml")
                                    else:
                                        print("    âŒ CLOUDY_PIXEL_PERCENTAGE not found in quality XML")
                                        continue
                                        
                                except ET.ParseError as xml_error:
                                    print(f"    âŒ Error parsing quality XML: {str(xml_error)[:50]}")
                                    continue
                                except Exception as xml_error:
                                    print(f"    âŒ Error reading quality XML: {str(xml_error)[:50]}")
                                    continue
                            else:
                                print(f"    âŒ L2A_QUALITY.xml not found in {os.path.join(granule_path, granule_id)}")
                                continue
                            
                            del cloud_raster
                            
                        except Exception as e:
                            print(f"Error: {str(e)[:30]}...")
                            continue
                    else:
                        print(f"    {granule_id[:15]}: No cloud mask found")
                
                # Calculate cloud coverage for this SAFE folder
                if safe_total_pixels > 0 and safe_tiles_analyzed > 0:
                    safe_cloud_coverage = (safe_cloud_pixels / safe_total_pixels) * 100
                    print(f"    ğŸ“ˆ SAFE folder cloud coverage: {safe_cloud_coverage:.1f}% ({safe_tiles_analyzed} tiles)")
                    
                    # Create new SAFE folder name with cloud coverage prepended for sorting (zero-padded, 2 decimal places)
                    cloud_pct_str = f"{safe_cloud_coverage:06.2f}".replace(".", "_")  # Zero-pad to 6 digits (XX_XX format)
                    new_safe_name = f"cloud{cloud_pct_str}pct_{safe}"  # Prepend cloud info to filename
                    new_safe_path = os.path.join(folder_path, new_safe_name)
                    
                    # Don't rename if cloud coverage is 0%
                    if safe_cloud_coverage == 0:
                        print(f"    â„¹ï¸  No clouds detected (0%) - folder name unchanged")
                        print(f"    ğŸŸ¢ Clean imagery ({safe_cloud_coverage:.1f}%)")
                    # Rename SAFE folder if name would change and has clouds
                    elif new_safe_name != safe:
                        try:
                            print(f"    ğŸ”„ Renaming SAFE folder to: {new_safe_name}")
                            
                            # Clear any potential file locks first
                            import gc
                            gc.collect()
                            
                            # Try the rename operation
                            os.rename(safe_folder_path, new_safe_path)
                            print(f"    âœ… Successfully renamed SAFE folder")
                            
                            # Report cloud coverage status
                            if safe_cloud_coverage <= max_cloud_coverage:
                                print(f"    ğŸŸ¢ Low cloud coverage ({safe_cloud_coverage:.1f}% â‰¤ {max_cloud_coverage}%)")
                            else:
                                print(f"    ğŸŸ¡ High cloud coverage ({safe_cloud_coverage:.1f}% > {max_cloud_coverage}%)")
                                
                        except PermissionError as perm_error:
                            print(f"    âš ï¸  Permission denied - folder may be in use. Cloud coverage: {safe_cloud_coverage:.1f}%")
                            print(f"    ğŸ’¡ Try closing any applications that might have files open in this folder")
                        except Exception as rename_error:
                            print(f"    âŒ Could not rename SAFE folder: {rename_error}")
                            print(f"    ğŸ“Š Cloud coverage recorded: {safe_cloud_coverage:.1f}%")
                    else:
                        print(f"    â„¹ï¸  SAFE folder name unchanged")
                else:
                    print(f"    âš ï¸  No valid cloud data found - SAFE folder name unchanged ({safe_tiles_analyzed} tiles)")

print("âœ… Cloud coverage analysis complete!")
```

```python
import os
import re

# -----------------------------
# Remove Cloud Percentage Suffixes from SAFE Folder Names
# -----------------------------

# Configuration - Edit this path as needed
base_directory = r"C:\Users\ss2596\Documents\SENTINEL2 Mufunta\Mufunta_Sentinel2_7_2025"

def remove_cloud_suffixes(directory_path):
    """
    Remove cloud percentage prefixes (cloudXX_Xpct_) and suffixes (_cloudXX_Xpct) from SAFE folder names
    """
    
    if not os.path.exists(directory_path):
        print(f"âŒ Directory not found: {directory_path}")
        return
    
    # Get all SAFE folders with cloud prefixes or suffixes
    safe_folders = [f for f in os.listdir(directory_path) if f.endswith(".SAFE")]
    cloud_folders = [f for f in safe_folders if ("cloud" in f and "pct" in f) or ("_cloud" in f and "pct" in f)]
    
    if not cloud_folders:
        print(f"No SAFE folders with cloud prefixes or suffixes found in: {directory_path}")
        return
    
    print(f"Found {len(cloud_folders)} SAFE folders with cloud prefixes/suffixes to rename:")
    
    renamed_count = 0
    failed_count = 0
    
    for folder in cloud_folders:
        # Use regex to remove the cloudXpct_ prefix and _cloudXpct suffix patterns
        # Handle both prepended format (cloudXX_XXpct_) and appended format (_cloudXX_XXpct)
        original_name = re.sub(r'^cloud\d+(?:_\d+)?pct_', '', folder)  # Remove prepended format (handles 1 or 2 decimal places)
        original_name = re.sub(r'_cloud\d+(?:_\d+)?pct', '', original_name)  # Remove appended format
        
        if original_name == folder:
            print(f"  âš ï¸  Skipping {folder} - no cloud prefix/suffix pattern found")
            continue
        
        current_path = os.path.join(directory_path, folder)
        new_path = os.path.join(directory_path, original_name)
        
        # Check if target name already exists
        if os.path.exists(new_path):
            print(f"  âŒ Cannot rename {folder} -> {original_name} (target already exists)")
            failed_count += 1
            continue
        
        try:
            print(f"  ğŸ”„ Renaming: {folder} -> {original_name}")
            
            # Clear any potential file locks
            import gc
            gc.collect()
            
            os.rename(current_path, new_path)
            print(f"  âœ… Successfully renamed")
            renamed_count += 1
            
        except PermissionError:
            print(f"  âŒ Permission denied - folder may be in use")
            failed_count += 1
        except Exception as e:
            print(f"  âŒ Error renaming folder: {str(e)[:50]}")
            failed_count += 1
    
    print(f"\nğŸ“Š Summary:")
    print(f"  âœ… Successfully renamed: {renamed_count} folders")
    print(f"  âŒ Failed to rename: {failed_count} folders")
    print(f"  ğŸ“ Total processed: {len(cloud_folders)} folders")

# Run the cleanup
print("ğŸ§¹ Starting cloud suffix removal...")
remove_cloud_suffixes(base_directory)
print("âœ… Cloud suffix removal complete!")
```
